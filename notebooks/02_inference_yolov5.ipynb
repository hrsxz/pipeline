```python
import os
import cv2
import matplotlib.pyplot as plt
from yolov5.utils.general import plot_one_box

# 定义图片与标签目录
image_dir = "yolov5/data/defect_detection/output"
label_dir = "yolov5/data/defect_detection/labels"

# 获取所有图片文件列表（以.jpg为例，如果有其他格式可适当修改或用glob）
image_files = [f for f in os.listdir(image_dir) if f.endswith(".jpg")]

for img_name in image_files:
    # 构建图像与标签文件路径
    img_path = os.path.join(image_dir, img_name)
    label_path = os.path.join(label_dir, img_name.replace(".jpg", ".txt"))

    # 读入图像
    img = cv2.imread(img_path)
    if img is None:
        print(f"Failed to read image {img_path}. Skipping...")
        continue

    img_h, img_w = img.shape[:2]  # 获取图像的宽高

    # 如果有对应的标签文件，则读取并在图像上绘制检测框
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                # YOLO label: class x_center y_center width height
                parts = line.strip().split()
                if len(parts) == 5:
                    class_id = int(parts[0])
                    x_center = float(parts[1])
                    y_center = float(parts[2])
                    width = float(parts[3])
                    height = float(parts[4])

                    # 将归一化坐标转换为像素坐标
                    x_center_pixel = x_center * img_w
                    y_center_pixel = y_center * img_h
                    w_pixel = width * img_w
                    h_pixel = height * img_h

                    x1 = int(x_center_pixel - w_pixel / 2)
                    y1 = int(y_center_pixel - h_pixel / 2)
                    x2 = int(x_center_pixel + w_pixel / 2)
                    y2 = int(y_center_pixel + h_pixel / 2)

                    bbox = [x1, y1, x2, y2]

                    # 在图像上绘制该标注框
                    plot_one_box(
                        x=bbox,
                        img=img,
                        color=(0, 0, 255),  # 白色框
                        label=f"class {class_id}",
                        line_thickness=1
                    )
    else:
        print(f"No label file for {img_name}, skipping bounding box drawing.")

    # 转换颜色并显示（在Notebook中显示）
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(2,2))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(img_name)
    plt.show()
```
